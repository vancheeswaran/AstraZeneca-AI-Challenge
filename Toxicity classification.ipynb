# TFID Vectorisation 

```python
import numpy as np
import sklearn as skl
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
import sklearn.metrics as skm
import matplotlib.pyplot as mpy
import seaborn as sn
```

```python
vz = TfidfVectorizer(stop_words='english', max_features=50000) 
X = vz.fit_transform(traindf)
col = ['id', 'target', 'comment_text', 'severe_toxicity', 'obscene',
       'identity_attack', 'insult', 'threat', 'asian', 'atheist', 'bisexual',
       'black', 'buddhist', 'christian', 'female', 'heterosexual', 'hindu',
       'homosexual_gay_or_lesbian', 'intellectual_or_learning_disability',
       'jewish', 'latino', 'male', 'muslim', 'other_disability',
       'other_gender', 'other_race_or_ethnicity', 'other_religion',
       'other_sexual_orientation', 'physical_disability',
       'psychiatric_or_mental_illness', 'transgender', 'white', 'created_date',
       'publication_id', 'parent_id', 'article_id', 'rating', 'funny', 'wow',
       'sad', 'likes', 'disagree', 'sexual_explicit',
       'identity_annotator_count', 'toxicity_annotator_count']
preds = np.zeros((testdf.shape[0], len(col))) 
AVGROC = [] 
```

# Model Training

```python
L=LogisticRegression()
```

```python
Y=traindf['target']
XTR,XTE,YTR,YTE=skl.model_selection.train_test_split(X,Y,test_size=0.3,random_state=30)
```

```python
L.fit(XTR,YTR)
```

```python
YP=L.predict(XTE)
```

# Performance 

```python
sn.heatmap(skm.confusion_matrix(YTE,YP),annot=True)
```

```python
skm.accuracy_score(YTE,YP)
```
